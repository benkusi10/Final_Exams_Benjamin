```python
#  Churn distribution by gender
gender_churn = dataset.groupby('Gender')['Churn Label'].value_counts().unstack()
gender_churn.plot(kind='bar', stacked=True)
plt.title('Churn Distribution by Gender')
plt.ylabel('Count')
plt.show()
```

```python
#  Churn rate by internet service type
service_churn = dataset.groupby('Internet Service')['Churn Label'].value_counts(normalize=True).unstack()
service_churn['Yes'].plot(kind='bar')
plt.title('Churn Rate by Internet Service Type')
plt.ylabel('Churn Rate')
plt.show()
```

```python
# Tenure distribution for churned customers
dataset[dataset['Churn Label'] == 'Yes']['Tenure Months'].plot(kind='hist', bins=20)
plt.title('Tenure Distribution for Churned Customers')
plt.xlabel('Tenure (months)')
plt.ylabel('Count')
plt.show()
```

```python
# Monthly vs Total charges for churned customers
churned = dataset[dataset['Churn Label'] == 'Yes']
plt.scatter(churned['Monthly Charges'], churned['Total Charges'])
plt.title('Monthly vs Total Charges for Churned Customers')
plt.xlabel('Monthly Charges')
plt.ylabel('Total Charges')
plt.show()
```

```python
# Top 10 churn reasons
top_reasons = dataset['Churn Reason'].value_counts().head(10)
top_reasons.plot(kind='barh')
plt.title('Top 10 Reasons for Churn')
plt.xlabel('Count')
plt.show()
```

```python
#Monthly charges vs Tenure for churned customers
plt.figure(figsize=(10, 6))
plt.scatter(
    churned['Tenure Months'], 
    churned['Monthly Charges'], 
    alpha=0.7, 
    color='red'
)
plt.title('Monthly Charges vs. Tenure for Churned Customers')
plt.xlabel('Tenure (Months)')
plt.ylabel('Monthly Charges ($)')
plt.grid(True)
plt.show()
```
